```python
import os 
import random
from PIL import Image
```

```python
# for ImageNet-V2, transform given wordnet-id to number 

with open('../resources/imagenet_1k_label_order.txt', 'r') as imagenet_v2_label_order_file:
    inet_v2_label_order = imagenet_v2_label_order_file.readlines()
     
```

```python
# define wnid of classes retrieved images belong 
class_a = 'n07714571'
class_b_l1 = 'n04286575'
class_b_c1 = 'n03991062'
class_b_c2 = 'n02206856'
class_b_c3 = 'n07715103'
class_r = 'n02100877'

# define dataset paths images should be retrieved
img_str_info = [(class_a, 'class_a_inet_1k', '../datasets/ImageNet1k/imagenet1k/ILSVRC/Data/CLS-LOC/val_sorted'),
(class_a, 'class_a_inet_c', '../datasets/ImagenetC/gaussian_blur/3'), 
(class_a, 'class_a_inet_r', '../datasets/ImagenetR_orig/imagenet-r'),
(class_a, 'class_a_inet_v2', '../datasets/imagenetv2-threshold0.7-format-val'),
(class_r, 'class_r_inet_1k', '../datasets/ImageNet1k/imagenet1k/ILSVRC/Data/CLS-LOC/val_sorted'),
(class_b_c1, 'class_b_c1_inet_1k', '../datasets/ImageNet1k/imagenet1k/ILSVRC/Data/CLS-LOC/val_sorted'), 
(class_b_c2, 'class_b_c2_inet_1k', '../datasets/ImageNet1k/imagenet1k/ILSVRC/Data/CLS-LOC/val_sorted'),
(class_b_c3, 'class_b_c3_inet_1k', '../datasets/ImageNet1k/imagenet1k/ILSVRC/Data/CLS-LOC/val_sorted'), 
(class_r, 'class_r_inet_1k', '../datasets/ImageNet1k/imagenet1k/ILSVRC/Data/CLS-LOC/val_sorted')]

random.seed(9)

for wnid, store_name, dset_path in img_str_info: 
    # in case INet-V2 dataset-folder must first be mapped to corresponding wnid
    if store_name == 'class_a_inet_v2':
        wnid = inet_v2_label_order.index(wnid +'\n')
        wnid = str(wnid)

    # get list of all images from this dataset with defined wnid 
    all_imgs_paths = os.listdir(dset_path + '/' + wnid)

    # select randomly 3 images from image-list
    for i in range(3):
        img_index = random.randint(0, len(all_imgs_paths) - 1)
        next_img_path = dset_path + '/' + wnid + '/' + all_imgs_paths[img_index]
        im = Image.open(next_img_path)
        im.resize((200, 200), resample = Image.BICUBIC)
        im.save('../example_pics/' + f'{store_name}_{i}.png')


```
